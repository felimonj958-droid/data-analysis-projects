## Aggregate data per user


````python
# Calculate the number of calls made by each user per month. Save the result.

calls_per_user = calls.groupby(['user_id', 'month']).size().reset_index()
calls_per_user.columns = ['user_id', 'month', 'num_calls']

print("calls per user per month:")
print(calls_per_user.head(10))
print(f"total records: {len(calls_per_user)}")
````

````python
# Calculate the amount of minutes spent by each user per month. Save the result.

minutes_per_user = calls.groupby(['user_id', 'month'])['duration'].sum().reset_index()
minutes_per_user.columns = ['user_id', 'month', 'total_minutes']

print("minutes per user per month:")
print(minutes_per_user.head(10))
print(minutes_per_user['total_minutes'].describe())
````

````python
# Calculate the number of messages sent by each user per month. Save the result.

messages_per_user = messages.groupby(['user_id', 'month']).size().reset_index()
messages_per_user.columns = ['user_id', 'month', 'total_messages']

print("messages per user per month:")
print(messages_per_user.head(10))
print(messages_per_user['total_messages'].describe())
````

````python
# Calculate the volume of internet traffic used by each user per month. Save the result.

internet_per_user = internet.groupby(['user_id', 'month'])['mb_used'].sum().reset_index()
internet_per_user.columns = ['user_id', 'month', 'total_mb']

print("internet usage (MB) per user per month:")
print(internet_per_user.head(10))
print(internet_per_user['total_mb'].describe())
````

<div class="alert alert-success"; style="border-left: 7px solid green">
<b>‚úÖ Reviewer's comment V2</b> 
    
Aggregating data complete.
</div>

````python
# Merge the data for calls, minutes, messages, internet based on user_id and month

user_activity = calls_per_user.merge(minutes_per_user, on=['user_id', 'month'], how='outer')
user_activity = user_activity.merge(messages_per_user, on=['user_id', 'month'], how='outer')
user_activity = user_activity.merge(internet_per_user, on=['user_id', 'month'], how='outer')

# replace NaN values with 0 for months users displayed no activity

user_activity = user_activity.fillna(0)

print("merged user activity data:")
print(user_activity.head(10))
print(f"\ntotal user-month records: {len(user_activity)}")
print(f"unique users: {user_activity['user_id'].nunique()}")
````

````python
# Add the plan information

user_plan = users[['user_id', 'plan']].copy()
user_activity = user_activity.merge(user_plan, on='user_id', how='left')

# user plan pricing details
user_activity = user_activity.merge(plans, left_on='plan', right_on='plan_name', how='left')

print("user activity and plan information:")
print(user_activity[['user_id', 'month', 'plan', 'num_calls', 'total_minutes', 'total_messages', 
                     'total_mb']].head(10))
````

<div class="alert alert-success"; style="border-left: 7px solid green">
<b>‚úÖ Reviewer's comment V2</b> 
    
Tables are merged correctly.
</div>

````python
# Calculate the monthly revenue for each user

# exceeding allowance charges

user_activity['exceeding_minutes'] = np.maximum(0, user_activity['total_minutes'] - user_activity['minutes_included'])
user_activity['exceeding_messages'] = np.maximum(0, user_activity['total_messages'] - user_activity['messages_included'])

# conversion of mb to gb and exceeding allowance charges

user_activity['total_gb'] = np.ceil(user_activity['total_mb'] / 1024)
user_activity['included_gb'] = user_activity['mb_per_month_included'] / 1024
user_activity['exceeding_gb'] = np.maximum(0, user_activity['total_gb'] - user_activity['included_gb'])

# revenue components

user_activity['revenue_minutes'] = user_activity['exceeding_minutes'] * user_activity['usd_per_minute']
user_activity['revenue_messages'] = user_activity['exceeding_messages'] * user_activity['usd_per_message']
user_activity['revenue_data'] = user_activity['exceeding_gb'] * user_activity['usd_per_gb']

# monthly revenue

user_activity['monthly_revenue'] = (user_activity['usd_monthly_pay'] + 
                                    user_activity['revenue_minutes'] +
                                    user_activity['revenue_messages'] +
                                    user_activity['revenue_data'])
print("monthly revenue:")
print(user_activity[['user_id', 'month', 'plan', 'monthly_revenue']].head(10))
print(user_activity['monthly_revenue'].describe())

# verify aggregation

print(f"total user-month records: {len(user_activity)}")
print(f"unique users: {user_activity['user_id'].nunique()}")
print("\naverage usage by plan:")
print(user_activity.groupby('plan')[['num_calls', 'total_minutes', 'total_messages',
                                     'total_gb']].mean())
print("\nrevenue by plan:")
print(user_activity.groupby('plan')['monthly_revenue'].describe())
````
